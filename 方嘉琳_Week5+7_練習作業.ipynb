````python
fruit1 = 'apple'
number1 = 2
total1 = number1
print('摘了', number1, '個', fruit1, '，總共摘了', total1, '個水果；')
fruit2 = 'pear'
number2 = 1
total2 = total1 + number2
print('摘了', number2, '個', fruit2, '，總共摘了', total2, '個水果；')
fruit3 = 'peach'
number3 = 3
total3 = total2 + number3
print('摘了', number3, '個', fruit3, '，總共摘了', total3, '個水果；')
fruit4 = 'banana'
number4 = 6
total4 = total3 + number4
print('摘了', number4, '個', fruit4, '，總共摘了', total4, '個水果；')
````
